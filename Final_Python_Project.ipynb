###**CENTRAL TENDENCY**

```python
# MEASURE OF CENTRAL TENDENCY
central_col = ["Tot_HH", "HH_Farm_Activities", "HH_Non_Farm_Activities"]
central_df = df[central_col]

mean_values = round(central_df.mean(),2)
print("Mean values:\n", mean_values)

median_values = central_df.median()
print("\nMedian values:\n",median_values)

mode_values = central_df.mode().iloc[0]
print("\nMode values:\n", mode_values)
```

```python
central_df.plot(kind="bar", figsize=(8,5))

plt.title("Measure of Central Tendency")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

### **VARIANCE AND STANDARD DEVIATION**

```python
# Variance and Standard Deviation
stastical_col = ["Tot_HH", "HH_Farm_Activities", "HH_Non_Farm_Activities"]
stastical_df = df[stastical_col]

variance_values = round(stastical_df.var(),2)
print("Variance values:\n", variance_values)

std_values = round(stastical_df.std(),2)
print("\n Standard Deviation values:\n", std_values)
```

### **SKEWNESS AND KURTOSIS**

```python
# Skewness and Kurtosis
shape_col = ["Tot_HH", "HH_Farm_Activities", "HH_Non_Farm_Activities"]
shape_df = df[shape_col]

skewness_values = round(shape_df.skew(),2)
print("Skewness Values are:\n", skewness_values)

kurtosis_values = round(shape_df.kurtosis(),2)
print("\n Kurtosis Values are:\n", kurtosis_values )
```

#### **Interpretation**

- **Uttar Pradesh** has the **highest number of households**, followed by **Bihar and West Bengal**.

- **Large states** such as **Maharashtra, Madhya Pradesh, and Tamil Nadu** also contribute significantly.

```python
#Histogram plot
plt.figure(figsize = (8,4))
sns.histplot(df["HH_Farm_Activities"], bins =30, kde = True, color = "blue")
plt.xlabel("Total Household", color = "red")
plt.ylabel("Frequency", color = "red")
plt.title("Histogram of Total Household")
plt.show()
```

#### **Interpretation**
The histogram shows how the number of households per village is distributed across the dataset.

-  **Most villages have relatively low household counts** — the majority of data is concentrated close to the left side (lower values).
-  There is a **long tail to the right**, meaning a small number of villages have **very high household counts** compared to the rest.
-  The shape is **right‑skewed** (positively skewed), showing that **high household values** are less common.
-  The high spike at lower values suggests that **most villages have fewer households**, while only a few villages are bigger in terms of household size.

- The **farming households distribution peaks more sharply at lower values**, indicating that agriculture remains the **primary livelihood for the majority of villages.**

- **Non-farming households exhibit a longer right tail**, suggesting greater variability and the presence of villages with **stronger livelihood diversification.**

- The substantial overlap between the two curves highlights that **farming and non-farming activities coexist within villages,** though their relative intensity varies with village size.

```python
#create new column
df["HH_Farm_percent"] = round((df["HH_Farm_Activities"] / df["Tot_HH"] * 100),2)
```

```python
#create new column
df["HH_Non_Farm_percent"] = round((df["HH_Non_Farm_Activities"] / df["Tot_HH"] * 100),2)
```

```python
infra_cols = [
    "Avail_Govt_Seed_Centres",
    "Avail_Warehouse_Food_Grain_Storage",
    "Avail_Custom_Hiring_Centre",
    "Avail_Community_Rain_Water_Harvesting_System",
    "Avail_Watershed_Dev_Proj",
    "Avail_Primary_Processing_Facilities"
]

df["Infra_Count"] = df[infra_cols].apply(
    lambda x: (x == "Yes").sum(),
    axis=1
)
```

```python
#classify the hh in lmh
df["Infra_Level"] = pd.cut(
    df["Infra_Count"],
    bins=[-1, 2, 5, 7],
    labels=["Low Infrastructure", "Medium Infrastructure", "High Infrastructure"]
)
```

```python
#Box Plot
plt.figure(figsize = (10,6))
sns.boxplot(
    x="Infra_Level",
    y="HH_Non_Farm_percent",

    data=df
)

plt.title("Non-Farm Household Share Across Infrastructure Levels")
plt.xlabel("Level of Agricultural Infrastructure", color ="red")
plt.ylabel("Non-Farm Households (%)", color ="red")
plt.show()
```

#### **Interpretation**
- Villages with **medium agricultural infrastructure show the highest median share of non-farm households**, indicating a more diversified livelihood structure at this level.

- **Low-infrastructure villages** have a **wide spread in non-farm household** share, suggesting **economic uncertainty and mixed livelihood dependence**.

- In **high-infrastructure villages**, the **median non-farm share slightly declines**, implying **stronger agricultural opportunities may reduce reliance** on non-farm activities.

- The presence of outliers across all infrastructure levels highlights **heterogeneity in livelihood choices**, influenced by **local economic and social factors** beyond infrastructure alone.

```python
#Heatmap
crosstab = pd.crosstab(
    df["Avail_Farmers_Collective"],
    df["Avail_Primary_Processing_Facilities"]
)

sns.heatmap(
    crosstab,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title("Farmers' Collectives vs Processing Facilities")
plt.xlabel("Processing Facilities Available", color ="red")
plt.ylabel("Farmers' Collective Available", color ="red")
plt.show()
```

#### **Interpretation**
- The **majority of villages** fall under the **“No farmers’ collective and no processing facilities”** category, indicating a significant infrastructure gap in rural agribusiness support.

- **Villages with both farmers’** collectives and processing facilities form a relatively small share, suggesting **limited integrated value-chain development.**

- Presence of **PACS or FPOs** slightly **increases the livelihood of having processing facilities**, highlighting the supportive role of collective institutions.

- Overall, the heatmap reveals a **strong association** between institutional organization and infrastructure availability, but with **substantial scope** for policy intervention to **improve coverage**.

###**3. Multivariate Analysis**

```python
#Scatter Plot
groups = df.groupby('Avail_Farmers_Collective')

for name, group in groups:
    plt.scatter(group['Tot_HH'], group['HH_Farm_Activities'], label=name)

plt.xlabel('Total Households', color = "red")
plt.ylabel('Farm Activity Households', color = "red")
plt.title('Impact of Farmers Collective')
plt.legend()
plt.show()
```

#### **Interpretation**
- Villages with farmers’ collectives **(FPOs, PACS, or both)** tend to show **higher farm-activity household counts**, suggesting a positive association between institutional support and agricultural participation.

- Villages **without any farmers’ collective** are more dispersed and often show **lower farm activity** **relative to village size**, indicating weaker agricultural engagement.

- The **presence of both FPOs and PACS** is associated with some of the **highest farm activity levels**, reflecting the benefits of **combined institutional mechanisms.**

- Overall, the scatter indicates that while village size matters, **collective-based organization** **enhances farm activity intensity**, supporting the role of farmer institutions in strengthening rural livelihoods.

```python
#Scatter plot
g = sns.lmplot(
    data=df,
    x= "Infra_Count",
    y= "HH_Non_Farm_percent",
    col = "Avail_Community_Rain_Water_Harvesting_System",
    scatter_kws={"alpha":0.3},
    height=5,
    aspect=1
)

g.set_axis_labels("Infrastructure Count", "Non-Farm Households (%)", color = "red")
g.fig.suptitle("Role of Water Infrastructure in Livelihood Shift", y=1.05)
plt.show()
```

#### **Interpretation**
- In villages **without rainwater harvesting**, as infrastructure increases, more **households move to non-farm activities**.

In villages **with rainwater harvesting**, the percentage of non-farm households **does not change much**, even when infrastructure increases.

This shows that **good water availability helps people continue farming**, while lack of water pushes households towards non-farm work.

In villages **without water support**, the impact of infrastructure on livelihoods **becomes weaker**, and more households shift to non-farm activities.

In villages **with rainwater harvesting**, farming remains stable, showing that **water acts as a multiplier that strengthens** the benefits of infrastructure.

```python
#Heatmap
num_cols = df[[
    "Tot_HH",
    "HH_Farm_Activities",
    "HH_Non_Farm_Activities"
]]

corr_matrix = num_cols.corr()

plt.figure(figsize=(6,4))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Between Household and Activity Variables")
plt.xticks( rotation=45)
plt.show()
```

#### **Interpretation**
- The heatmap highlights **positive relationships among household** count, farm activities, and non-farm activities.

- Total households show a **strong association** with participation in **both activity types.**

- Farm and non-farm activities are only moderately linked, **indicating varied livelihood choices.**

- The pattern suggests households distribute their involvement across **different economic activities** rather than relying on one.
